<h1>자료구조 - 정렬</h1>
<p>정렬<br> : 값의 대소관계에 따라 데이터 집합을 오름/내림차순으로 배치하는 것</p>

<p>sraight 단순 선형만 다룸</p>

<h4>정렬 종류</h4>
1. 버블 정렬<br>
 : 이웃한 원소끼리 비교, 필요 시 교환하기 반복하는 작업방식 (= 단순교환정렬)<br>
2. 선택 정렬<br>
 : 가장 작은 원소부터 선택해 알맞은 위치로 옮기기(=교환하기) 반복하는 작업방식 (=단순선택정렬)<br>
3. 삽입 정렬<br>
 : 타깃 원소보다 더 앞쪽에서 알맞은 위치로 삽입하기 반복하는 작업방식<br>
4. 퀵 정렬<br>
 : 임의의 피봇(기준타깃) 잡고 한쪽엔 작은애들, 한쪽엔 큰애들 넣길 반복하는 작업방식<br>
5. 병합 정렬<br>
 : 배열을 앞-뒤 두부분으로 나눠 각각 정렬 후 병합하는 작업방식<br>
6. 힙 정렬<br>
 : 가장 아랫부분의 작은 서브트립터 상향식으로 진행하여 건체 배열을 힙으로 만드는 작업방식<br>
	<small>* 힙 heap : 부모 값이 자식 값보다 항상 크다는 조건 만족하는 "완전 이진 트리"</small><br>
